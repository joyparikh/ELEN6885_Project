```python

returns = {i:[0,0] for i in range(9)}

with open('results.txt', 'r') as file:
    for line in file:
        if line[:2] == 'sc':
            split = line.split()
            returns[int(split[2][:-1])].append(float(split[-1]))

for i in range(1,9):
    returns[i] = returns[i][1:]

            
```

```python
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(14,5))
for scramble in [1,2,3,4, 5]:

    plt.plot([x*5000 for x in range(len(returns[scramble])-1)], returns[scramble][1:], label=f'scramble_max = {scramble}')
plt.grid()
plt.legend()
plt.title('DQN efficacy with (1000,1000,1000,300) hidden layer architecture, learning rate decay within (1e-3, 1e-5)')
plt.xlabel('MC Step Count')
plt.ylabel('Solve Accuracy (for given scramble_max)')
plt.show
```
